### Big Data:

**Definition**:
Massive and complex datasets that cannot be handled by traditional data management tools due to the inability to scale effectively for data collection, storage, and processing.

**Key Characteristics**(known as 5Vs):
1. **Volume**: Scale of data, typically in the order of TeraBytes or larger
2. **Variety**: Different types of data; Diversity of data; Different Sources.
3. **Velocity**: The speed the data is generated and accumulated and a never-stop process often in real-time.
4. **Veracity**: Quality and accuracy of data such as Consistency, Completeness, Integrity and ambiguity.
5. **Value**: The huge potential insights and benefits contained in the data.

## Introduction to Data Science

**Definitions**: Data science is an interdisciplinary study that applies
- Math and Statistics
- Programming
- Domain-Specific Knowledge
  
to extract valuable and meaningful insights from data for decision-making.

**Ethics in Data Science**
- **Legal Compliance**: Ensure data is sourced from legal, trustworthy origins and adhere to with law and regulations during data usage.
- **Privacy**: Safeguard the privacy and confidentiality of individuals' data.
- **Bias and Fairness**: Identifying and mitigating biases in data and algorithms to prevent unfair treatment or discrimination.
- **Transparency**: Being transparent about data sources, methodologies, and algorithms used in data analysis.
- **Security**: Protect data from unauthorized access and breaches


**Data Science Workflow:**
1. **Define the Problem**: Understand the problem you are trying to solve. Clearly define the problem you aim to solve and establish the specific and measurable objectives.
2. **Data Collection**: Gather relevant data from reliable and legal sources and select data that aligns with the problems.
3. **Data Exploration**: Conduct exploratory analysis to understand the characteristics of data. For example, get a high level understanding of data by summary statistics; Use visual tools (like box plot, histogram, line chart) to explore the distribution, patterns and trends within the data; Identify potential issues such as missing values, duplicates, outliers and inconsistent data types.
4. **Data Processing**: Preprocess the data to ensure its quality and consistency. For example, address issues identified during exploration; normalize and standardize data, transform categorical variables into numerical formats, reduce the number of attributes with a acceptable loss in information(explained in detail later).
5. **Modeling**: Select the appropriate statistical model. Split data into train and test sets. Train the model on the training data. Evaluate the model performance on test data.
6. **Insights, Interpretation and Communication**: Interpret the model to derive actionable insights. Prepare clear, concise reports and visualizations. Think about how to communicate your results with your audience.

#### Broadcasting
Broadcasting work with arrays of different shapes during arithmetic operations. It is a way of stretching the smaller array across the larger array so they have compatible shapes for element-wise operations.

````python
a = np.array([1, 2, 3])
b = 2
c = np.array([
    [1], 
    [2], 
    [3]])
result1 = a + b # The scalar b is broadcasted to the shape of a and behaves like an array [2, 2, 2].
result2 = a + c # a has shape (1, 3), b has shape (3, 1), The array ais broadcast to shape (3, 3) to match b


print("Broadcasting Result(a+b):\n", result1,
      "\nBroadcasting Result(a+c):\n", result2)
````

### Matrix and Its Eigenvalue Decomposition (EVD)

Given a square Matrix A :


$$A = \begin{bmatrix} 
1 & 2 \\ 
3 & 4 
\end{bmatrix}$$


Eigenvalue Decomposition (EVD):

The Eigenvalue Decomposition of matrix A is given by:

$$ A = U \Sigma U^{-1} $$


where:
- $U$  is the matrix of eigenvectors,
- $\Sigma$  is the diagonal matrix of eigenvalues,
- $U^{-1}$  is the inverse of the matrix  $U$ .


Matrix U (Matrix of Eigenvectors):


$$U = \begin{bmatrix} 
-0.82456484 & -0.41597356 \\ 
0.56576746 & -0.90937671 
\end{bmatrix}$$


Matrix $\Sigma$ (Diagonal Matrix of Eigenvalues):


$$\Sigma = \begin{bmatrix} 
-0.37228132 & 0 \\ 
0 & 5.37228132
\end{bmatrix}$$


Matrix $U^{-1}$  (Inverse of Matrix U):


$$U^{-1} = \begin{bmatrix} 
-0.92305231 &  0.42222915 \\ 
-0.57427572 & -0.83696501
\end{bmatrix}$$

Matrix Multiplication Verification:

Multiplying these matrices should reconstruct the original matrix A:

 $$A = U \Sigma U^{-1} $$

Performing the multiplication:

$$ U \Sigma U^{-1} = \begin{bmatrix} 
-0.82456484 & -0.41597356 \\ 
0.56576746 & -0.90937671 
\end{bmatrix} \times \begin{bmatrix} 
-0.37228132 & 0 \\ 
0 & 5.37228132
\end{bmatrix} \times \begin{bmatrix} 
-0.92305231 &  0.42222915 \\ 
-0.57427572 & -0.83696501
\end{bmatrix} $$

This should equal:

\begin{bmatrix} 
1 & 2 \\ 
3 & 4 
\end{bmatrix}

### Matrix and Its Singular Value Decomposition (SVD)
You find the content below is highly similar to Eigenvalue Decomposition above, but EVD is only applicabl to square matrix if it is diagonalizable, but not all matrices are diagonalizable. SVD generalises to any matrix which implies that it does not require the matrix to be square or diagonalizable.

Given a Matrix A :

$$ A = \begin{bmatrix} 
1 & 2 \\ 
3 & 4 
\end{bmatrix}$$

Singular Value Decomposition (SVD):

The SVD of matrix A is given by:

$$ A = U \Sigma V^T $$

where:

- U is the matrix of left singular vectors,
- $\Sigma$ is the diagonal matrix of singular values,
- $V^T$ is the matrix of right singular vectors transposed.

Matrix U:

$$ U = \begin{bmatrix} 
-0.40455358 &  -0.9145143 \\ 
-0.9145143  &  0.40455358 
\end{bmatrix} $$

Matrix $\Sigma$ (Diagonal Matrix of Singular Values):

$$ \Sigma = \begin{bmatrix} 
5.4649857 & 0 \\ 
0 & 0.36596619 
\end{bmatrix} $$

Matrix $V^T$:

$$ V^T = \begin{bmatrix} 
-0.57604844 & -0.81741556 \\ 
0.81741556 & -0.57604844 
\end{bmatrix} $$

Matrix Multiplication Verification:

Multiplying these matrices should reconstruct the original matrix A:

 $$A = U \Sigma V^T $$

Performing the multiplication:

$$ U \Sigma V^T = \begin{bmatrix} 
-0.40455358 &  -0.9145143 \\ 
-0.9145143  &  0.40455358 
\end{bmatrix} \times \begin{bmatrix} 
5.4649857 & 0 \\ 
0 & 0.36596619 
\end{bmatrix} \times \begin{bmatrix} 
-0.57604844 & -0.81741556 \\ 
0.81741556 & -0.57604844 
\end{bmatrix} $$

This should equal:

\begin{bmatrix} 
1 & 2 \\ 
3 & 4 
\end{bmatrix} 


````python
# Singular Value Decomposition

A = np.array([[1, 2],
              [3, 4],
              ])

# Perform Singular Value Decomposition
U, S, Vt = np.linalg.svd(A)

# Convert S to a diagonal matrix
Sigma = np.diag(S)

# Display the results
print("Matrix A:")
print(A)

print("\n matrix U:")
print(U)

print("\nSigma (Diagonal matrix of singular values):")
print(Sigma)

print("\n matrix V^T:")
print(Vt)

# Verify the reconstruction
A_reconstructed = U @ Sigma @ Vt
print("\nReconstructed Matrix A (U * Sigma * V^T):")
print(A_reconstructed)
````

### Fourier Transform 

The Fourier Transform is a powerful mathematical technique that transforms a time-domain signal into its frequency-domain representation. It is defined as follows:

$$ 
\mathcal{F}(f(t)) = F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-i \omega t} \, dt
$$ 

where:
- $f(t)$ is the original signal in the time domain,
- $F(\omega)$ is the transformed signal in the frequency domain,
- $ \omega$ is the angular frequency, and
- $ i $ is the imaginary unit.

The Fourier Transform decomposes a time-domain signal into its constituent frequencies, allowing us to analyze the frequency components within the signal. This is particularly useful in applications involving signal processing, as it enables operations such as filtering, signal compression, and analysis of periodicities.

In practical applications, the **Fast Fourier Transform (FFT)** is an efficient algorithm for computing the discrete Fourier Transform (DFT), which is the sampled version of the Fourier Transform. The FFT is widely used for analyzing frequency components, filtering signals, and detecting patterns in time-series data. By converting time-series data from the time domain to the frequency domain, the FFT helps reveal the **frequencies** that characterize the data, enabling better insights and analysis.

In summary, the Fourier Transform captures the frequency information of a time-series signal, making it an essential tool in fields such as engineering, physics, and data science.

````python
import matplotlib.pyplot as plt

# Create a sample signal (e.g., a sinusoidal signal)
Fs = 1000  # Sampling frequency (samples per second)
T = 1 / Fs  # Sampling interval (seconds)
L = 2000  # Length of the signal
t = np.arange(0, L) * T  # Time vector

# Signal with two frequencies
f1 = 50  # Frequency of the first component (Hz)
f2 = 120  # Frequency of the second component (Hz)
signal = 0.7 * np.sin(2 * np.pi * f1 * t) + 1 * np.sin(2 * np.pi * f2 * t)

# Compute the Fourier Transform
Y = np.fft.fft(signal)

# Compute the corresponding frequencies
frequencies = np.fft.fftfreq(len(signal), T)

# Plot the signal and its Fourier Transform
plt.figure(figsize=(12, 6))

# Plot the original signal
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# Plot the magnitude spectrum of the Fourier Transform
plt.subplot(2, 1, 2)
plt.plot(frequencies[:L // 2], np.abs(Y[:L // 2]))
plt.title('Magnitude Spectrum')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt

# Create a sample signal (same as before)
Fs = 1000
T = 1 / Fs
L = 1000
t = np.arange(0, L) * T
f1 = 50
f2 = 120
signal = 0.7 * np.sin(2 * np.pi * f1 * t) + np.sin(2 * np.pi * f2 * t)

# Compute the Fourier Transform
Y = np.fft.fft(signal)

# Compute the inverse Fourier Transform
reconstructed_signal = np.fft.ifft(Y)

# Plot the original and reconstructed signals
plt.figure(figsize=(12, 6))

# Plot the original signal
plt.subplot(2, 1, 1)
plt.plot(t, signal)
plt.title('Original Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

# Plot the reconstructed signal
plt.subplot(2, 1, 2)
plt.plot(t, np.real(reconstructed_signal))  # Only plot the real part
plt.title('Reconstructed Signal from FFT')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()
````
