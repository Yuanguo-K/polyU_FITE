### Pandas Highlights
**Core Data Structures in Pandas**
- Series: A one-dimensional labeled array that can hold any data type, similar to a column in spreadsheet.
- DataFrame: The core of Pandas, designed for structured data manipulation. It organizes data in a two-dimensional, table-like format, similar to a spreadsheet, allowing for intuitive data analysis and manipulation.

**Built on Top of NumPy**
- Efficient Data Computation and Storage: Pandas leverages NumPy's computational power and memory-efficient data storage, allowing for fast and complex operations on large datasets.

**Performance and Scalability**
- Scalable Data Operations: Pandas can handle operations on large datasets that would overwhelm Excel, such as merging, joining, and processing extensive data.
- Optimized for Performance: Offers high-performance computations, particularly for data-heavy tasks, ensuring that operations are both fast and scalable.


**Flexible Data Format**
- Data Format Support: DataFrame can be created manually by entering from scratch or importing from various data formats, and exported just as easily

**Efficient Data Handling**
- Complex Wrangling Support: Perform advanced tasks like filtering, merging, and aggregation with ease. Pandas offers powerful handling of missing data with features like filling, interpolation, duplicate removal etc.

**Seamless integration**
- Library Integration: Works smoothly with Python libraries such as NumPy, Matplotlib, Scikit-Learn, Datetime and etc.
- Database Interaction: Pandas provides robust functionality for connecting to and working with SQL databases, enabling seamless data extraction, manipulation, and storage.

#### Scraping the Web

Web scraping is a powerful technique for extracting data from websites by parsing the HTML content, which can be particularly useful in data science. This is useful for gathering information that is not readily available through APIs. Before diving into web scraping, it will be great if you meet HTML before, as it forms the backbone of most web pages. HTML (HyperText Markup Language) is an very structured language responsible for displaying the content on the web, allowing you to navigate and extract specific elements such as tables, lists, or paragraphs.
Understanding the structure of HTML documents will enable you to navigate through the tags and extract the data relevant to your analysis. For example, by identifying specific HTML tags (like `<table>`, Table Header `<th>`, Table Row `<tr>`), tables can be pulled out quickly with few lines of code

```python
import pandas as pd
import requests

# URL to fetch the flight data
url = "https://www.hongkongairport.com/flightinfo-rest/rest/flights/past?date=2024-11-18&lang=en&cargo=false&arrival=false"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Load the JSON response
    data = response.json()

    # Initialize an empty list to hold flight information
    flights = []
    # Iterate through the data to normalize and store relevant information
    for day in data:
        
        # for flight in day["list"]:
        #     print(day)
        #     print('\n')
        #     print(flight)
        is_arrival = day['arrival']
        for flight_info in day["list"]:
            shared = len(flight_info['flight']) > 1
            n = 0
            for flight in flight_info['flight']:
                flights.append({
                    "Date": day["date"],
                    'flight_type':'inbound' if is_arrival else 'outbound',
                    "Time": flight_info["time"],
                    "Flight Number": flight['no'],
                    "Airline": flight["airline"],
                    "Status": flight_info["status"],
                    "Origin or Destination": flight_info["origin"] if is_arrival else flight_info['destination'],
                    "Terminal": flight_info["terminal"],
                    "Aisle": flight_info["aisle"],
                    "Gate": flight_info["gate"],
                    "Shared": "True" + str(n) if shared else "False"
                })
                n += 1

    # # Create a DataFrame from the list of flights
    df = pd.DataFrame(flights)

    # Print the DataFrame to see the results
    print(df)
else:
    print(f"Error: Unable to fetch data. Status Code {response.status_code}")
```

```python
# random walk

# Set the number of steps and the number of random walks
num_steps = 100
num_walks = 5

# Generate random steps: +1 or -1
steps = np.random.choice([-1, 1], size=(num_steps, num_walks))

# Create cumulative sum to simulate the random walk
random_walks = pd.DataFrame(steps, index=pd.date_range("01/09/2024", periods=num_steps), columns=range(1,num_walks+1)).cumsum()

# Plot the random walks
random_walks.plot(figsize=(10, 6), title='Random Walks')
plt.ylabel('Position')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for walk in random_walks.columns:
    plt.scatter(random_walks.index, random_walks[walk], s=5, label=f'Walk {walk}')

plt.title('Random Walks (Scatter Plot)')
plt.xlabel('Date')
plt.ylabel('Position')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2889. Reshape Data: Pivot

weather = {"headers":{"weather":["city","month","temperature"]},"rows":{"weather":[["Jacksonville","January",13],["Jacksonville","February",23],["Jacksonville","March",38],["Jacksonville","April",5],["Jacksonville","May",34],["ElPaso","January",20],["ElPaso","February",6],["ElPaso","March",26],["ElPaso","April",2],["ElPaso","May",43]]}}

headers = weather["headers"]["weather"]
rows = weather["rows"]["weather"]
weather = pd.DataFrame(rows, columns=headers)
print(weather, "\n")
weather = weather.pivot(index='month', columns='city', values='temperature')
print("\n\nVisualized as ")
weather
```

```python
# 2890. Reshape Data: Melt

report = {"headers":{"report":["product","quarter_1","quarter_2","quarter_3","quarter_4"]},"rows":{"report":[["Umbrella",417,224,379,611],["SleepingBag",800,936,93,875]]}}
headers = report["headers"]["report"]
rows = report["rows"]["report"]
report= pd.DataFrame(rows, columns=headers)
print(report)
report.melt(
        id_vars=["product"],
        value_vars=["quarter_1", "quarter_2", "quarter_3", "quarter_4"],
        var_name="quarter",
        value_name="sales"
)
```
