## **Statistical Models and Inference**

### **Regression**

Regression analysis is a fundamental statistical technique used to explore the relationship between a **dependent variable** (e.g., house price) and one or more **independent variables** (e.g., house area and age). It helps quantify how changes in the independent variables influence the dependent variable.

In regression, we estimate the conditional distribution $𝑝(𝑦_∗∣x_*)$ , which represents the probability distribution of the unknown response $y_*$ associated with a new test input $x_*$. This distribution serves multiple purposes.

- Prediction: Enables forecasting or estimating the value of $y$ for new input data.
- Interpretation: Provides insights into how the independent variables influence $y$. **Note**: Regression does not imply causation.


---

#### **Regression Techniques**

There are various approaches to regression analysis, with two primary techniques being:

- Ordinary Least Squares (OLS) – Minimizes the sum of squared errors between observed and predicted values.
- Gradient Descent – An optimization algorithm that iteratively updates model parameters to minimize error.
---


#### Analytical: Ordinary Least Square

Ordinary Least Square (OLS) is an analytical method to find the best-fitting line (regression line) that minimizes the sum of the squares of the vertical deviations from each data point to the line.

OLS and Maximum Likelihood Estimation (MLE) lead to same coefficient estimates if 
- The error term $\epsilon$ is normally distributed with same variance.
- The errors are indepedent of each other.

---
At the following example, only one dependent variable will be used fom simplicity
##### Linear Regression Model

The simple linear regression model can be expressed as:

$$  y_i = \beta_0 + \beta_1 x_i + \epsilon  $$

where:
-  $y_i$ is the i-th dependent variable.
-  $x_i$ is the i-th independent variable.
-  $\beta_0$ is the intercept.
-  $ \beta_1 $ is the slope.
-  $ \epsilon$ is the noise term drawn from standard normal distribution.

or in vector form for all data

$$
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{bmatrix} =
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_n
\end{bmatrix}
\begin{bmatrix}
\beta_0 \\
\beta_1
\end{bmatrix}
+
\begin{bmatrix}
\epsilon_1 \\
\epsilon_2 \\
\vdots \\
\epsilon_n
\end{bmatrix}
$$

$$
\mathbf{Y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}
$$
where:
- $\mathbf{Y}$ is the vector including all observed values.
- $\mathbf{X}$ is the matrix of all independent variables (including a column of ones for the intercept).
- $\boldsymbol{\beta}$ is the vector of coefficients.
- $\boldsymbol{\epsilon}$ is the vector of all error terms.

Please bear in mind X can increase in dimension horizontally to include more features, $\boldsymbol{\beta}$ can increase in dimension vertically, so can be expanded into multi-factor regression.

---
##### Ordinary Least Squares (OLS) Solution

To find the optimal values of $\beta_0$ and $\beta_1$, OLS minimizes the cost function, also known as the sum of squared residuals (SSR):

$$ \text{SSR} = \sum_{i=1}^{n} (y_i - \hat{y_i})^2 $$

where:
- $ y_i $ is the actual value.
- $ \hat{y_i}$ is the predicted value $ \hat{y_i} = \beta_0 + \beta_1 x_i $.


The formulas to compute $\beta_0$ and $\beta_1$ are:

$$ \beta_0 =  \frac{\sum_{i=1}^{n}{y_i}}{n} - \beta_1 \frac{\sum_{i=1}^{n}{x_i}}{n} = \bar{y} - \beta_1 \bar{x} $$


$$ \beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2} $$


where:
- $ \bar{x} $ is the mean of the independent variable.
- $ \bar{y} $ is the mean of the dependent variable.


**Vector** form solution:


$$
\hat{\boldsymbol{\beta}} = \arg\min_{\boldsymbol{\beta}} (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta})
$$



$$
\hat{\boldsymbol{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}
$$

---

**Derivation**:

To find the optimal values of $\beta_0$ and $\beta_1$ in Ordinary Least Squares (OLS) linear regression, we need to minimize the sum of squared residuals (SSR). The model is:


$$ \hat{y}_i = \beta_0 + \beta_1 x_i $$ 


The cost function (SSR) is:


$$ \text{SSR} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 = \sum_{i=1}^{n} (y_i - (\beta_0 + \beta_1 x_i))^2 $$


To find the optimal $\beta_0$ and $\beta_1$, we take partial derivatives of SSR with respect to $\beta_0$ and $\beta_1$ and set them to zero.

**Step 1**: Partial Derivative with respect to $\beta_0$}


$$\frac{\partial \text{SSR}}{\partial \beta_0} = \frac{\partial}{\partial \beta_0} \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i)^2$$


Using the chain rule:


$$\frac{\partial \text{SSR}}{\partial \beta_0} = \sum_{i=1}^{n} 2(y_i - \beta_0 - \beta_1 x_i)(-1)$$


Simplifying:


$$\frac{\partial \text{SSR}}{\partial \beta_0} = -2 \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i)$$


Setting the derivative to zero:


$$-2 \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) = 0$$



$$\sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) = 0$$



$$\sum_{i=1}^{n} y_i - n\beta_0 - \beta_1 \sum_{i=1}^{n} x_i = 0$$


Solving for $\beta_0$:


$$n\beta_0 = \sum_{i=1}^{n} y_i - \beta_1 \sum_{i=1}^{n} x_i$$



$$\beta_0 = \frac{1}{n} \sum_{i=1}^{n} y_i - \beta_1 \frac{1}{n} \sum_{i=1}^{n} x_i$$


Let $\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i$ and $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$, then:


$$\beta_0 = \bar{y} - \beta_1 \bar{x}$$


**Step 2**: Partial Derivative with respect to $\beta_1$


$$\frac{\partial \text{SSR}}{\partial \beta_1} = \frac{\partial}{\partial \beta_1} \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i)^2$$


Using the chain rule:


$$\frac{\partial \text{SSR}}{\partial \beta_1} = \sum_{i=1}^{n} 2(y_i - \beta_0 - \beta_1 x_i)(-x_i)$$


Simplifying:


$$\frac{\partial \text{SSR}}{\partial \beta_1} = -2 \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) x_i$$


Setting the derivative to zero:


$$-2 \sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) x_i = 0$$


$$\sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) x_i = 0$$


Substitute $\beta_0 = \bar{y} - \beta_1 \bar{x}$ into the equation:


$$\sum_{i=1}^{n} (y_i - (\bar{y} - \beta_1 \bar{x}) - \beta_1 x_i) x_i = 0$$



$$\sum_{i=1}^{n} (y_i - \bar{y} + \beta_1 \bar{x} - \beta_1 x_i) x_i = 0$$



$$\sum_{i=1}^{n} (y_i - \bar{y}) x_i + \beta_1 \sum_{i=1}^{n} (\bar{x} - x_i) x_i = 0$$


Since $\sum_{i=1}^{n} (\bar{x} - x_i) x_i = \bar{x} \sum_{i=1}^{n} x_i - \sum_{i=1}^{n} x_i^2$:


$$\sum_{i=1}^{n} (y_i - \bar{y}) x_i + \beta_1 (\bar{x} \sum_{i=1}^{n} x_i - \sum_{i=1}^{n} x_i^2) = 0$$



$$\sum_{i=1}^{n} (y_i - \bar{y}) x_i = \beta_1 \sum_{i=1}^{n} (x_i - \bar{x}) x_i$$



$$\sum_{i=1}^{n} (y_i - \bar{y}) x_i = \beta_1 \sum_{i=1}^{n} (x_i - \bar{x})^2$$


Solving for $\beta_1$:


$$\beta_1 = \frac{\sum_{i=1}^{n} (y_i - \bar{y}) (x_i - \bar{x})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$



The optimal values of \beta_0 and \beta_1 are:


$$\beta_1 = \frac{\sum_{i=1}^{n} (y_i - \bar{y}) (x_i - \bar{x})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}$$



$$\beta_0 = \bar{y} - \beta_1 \bar{x}$$


These values minimize the sum of squared residuals (SSR) in the linear regression model.

---

**Vector** form solution:
To find the optimal $\hat{\boldsymbol{\beta}}$, we take the derivative with respect to $\boldsymbol{\beta}$  and set it to zero:

$$
\frac{\partial}{\partial \boldsymbol{\beta}} (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta})^T (\mathbf{Y} - \mathbf{X} \boldsymbol{\beta}) = 0
$$

Solving this gives the normal equation:

$$
\mathbf{X}^T \mathbf{Y} = \mathbf{X}^T \mathbf{X} \hat{\boldsymbol{\beta}}
$$

Therefore, the OLS estimator is:
$$\hat{\boldsymbol{\beta}} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}$$
---

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# 1. Synthetically generate the data
# Generate random beta values
beta_0 = np.random.randn() * 10
beta_1 = np.random.randn() * 10

# Generate random x values
n = 100  # Number of data points
x = 2 * np.random.uniform(-10, 10, n).reshape(-1, 1)  # Ensure x is a column vector

# Generate y values using the linear regression model with added noise
epsilon = np.random.randn(n, 1) * 10
y = beta_0 + beta_1 * x + epsilon

# 2. Analytical calculation (Ordinary Least Squares)
# 2.1 
# Calculate means of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculate beta_1 using the provided equation
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
beta_1_hat = numerator / denominator

# Calculate beta_0 using the provided equation
beta_0_hat = y_mean - beta_1_hat * x_mean
print(beta_0_hat, beta_1_hat)
print(f"OLS Beta_0: {beta_0_hat}, OLS Beta_1: {beta_1_hat}")

# 2.2
x_b = np.column_stack([np.ones((n, 1)), x])  # Add x_0 = 1 to each instance
theta_best = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)

# Extracting the coefficients
beta_0_ols, beta_1_ols = theta_best
print(f"OLS Beta_0: {beta_0_ols[0]}, OLS Beta_1: {beta_1_ols[0]}")

# Plot the data and the regression lines
plt.scatter(x, y, color='blue')
plt.plot(x, beta_0 + beta_1 * x, "r", linewidth=2, label="Generated Data Line")
plt.plot(x, beta_0_hat + beta_1_hat * x, "black", linewidth=3, label="OLS Regression Line")
plt.plot(x, beta_0_ols + beta_1_ols * x, "grey", linewidth=0.5, label="Matrix-Form Solution Line")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Comparison of Regression Lines')
plt.show()
```

#### Computational: Gradient Descent

Gradient Descent is a computational optimization algorithm used to minimize the cost function by iteratively updating the parameters of the model.



---
##### Gradient Descent Algorithm

1. Initialize the parameters $\beta_0$ and $\beta_1$ with random values.
2. Compute the predicted value $\hat{y_i} = \beta_0 + \beta_1 x_i$.
3. Calculate the cost function (Mean Squared Error, MSE):

$$ J(\beta_0, \beta_1) = \frac{1}{2n} \sum_{i=1}^{n} (\hat{y_i} - y_i)^2 $$

4. Update the parameters using the gradients of the cost function:

$$  \beta_0 := \beta_0 - \alpha \frac{\partial J}{\partial \beta_0} $$

where:
- $\alpha$ is the learning rate.
- $\frac{\partial J}{\partial \beta_1}$ and $\frac{\partial J}{\partial \beta_0}$ are the partial derivatives of the cost function with respect to $\beta_1$ and $\beta_0$, respectively.

The partial derivatives are calculated as:

$$ \frac{\partial J}{\partial \beta_1} = \frac{1}{n} \sum_{i=1}^{n} (\hat{y_i} - y_i) x_i $$

$$ \frac{\partial J}{\partial \beta_0} = \frac{1}{n} \sum_{i=1}^{n} (\hat{y_i} - y_i) $$

5. Repeat steps 2-4 until convergence (i.e., when the change in the cost function is below a predefined threshold).

By using these methods, we can find the optimal solution for the parameters $\beta_0$ and $\beta_1$ to best fit the data in a linear regression model.

#### Conclusion

Both Ordinary Least Squares (OLS) and Gradient Descent are fundamental techniques for performing linear regression. OLS provides an analytical solution that is efficient for smaller datasets, while Gradient Descent is a flexible optimization method that can handle larger datasets and more complex models.

---


```python
# 3. Gradient Descent
def compute_cost(theta, x, y):
    m = len(y)
    predictions = x.dot(theta)
    cost = (1/2*m) * np.sum(np.square(predictions - y))
    return cost

def gradient_descent(x, y, theta, learning_rate, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    theta_history = np.zeros((iterations, 2))

    for i in range(iterations):
        predictions = x.dot(theta)
        theta = theta - (1/m) * learning_rate * (x.T.dot((predictions - y)))
        theta_history[i,:] = theta.T
        cost_history[i] = compute_cost(theta, x, y)
    
    return theta, cost_history, theta_history

# Initialize parameters for gradient descent
theta = np.random.randn(2, 1)
iterations = 100
learning_rate = 0.01

# Run gradient descent
x_b = np.c_[np.ones((n, 1)), x]  # Add x_0 = 1 to each instance
theta_gd, cost_history, theta_history = gradient_descent(x_b, y, theta, learning_rate, iterations)

# Extracting the coefficients
beta_0_gd, beta_1_gd = theta_gd
print(f"Original Beta_0: {beta_0}, Original Beta_1:{beta_1}  \n OLS Beta_0: {beta_0_ols}, OLS Beta_1: {beta_1_ols} \nGradient Descent Beta_0: {beta_0_gd[0]}, Gradient Descent Beta_1: {beta_1_gd[0]}")

# # Plot the cost history
# plt.plot(range(iterations), cost_history, 'b.')
# plt.xlabel('Iterations')
# plt.ylabel('Cost')
# plt.title('Cost History')
# plt.show()

# Plot the data and the regression lines
plt.scatter(x, y, color='blue')
plt.plot(x, beta_0 + beta_1 * x, "r-", linewidth=2, label="Generated Data Line")
plt.plot(x, beta_0_ols + beta_1_ols * x, "g-", linewidth=2, label="OLS Regression Line")
plt.plot(x, beta_0_gd + beta_1_gd * x, "y-", linewidth=2, label="Gradient Descent Line")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Comparison of Regression Lines')
plt.show()
```

```python
# Function to remove outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Remove outliers in "area" and "price"
df = remove_outliers(df, 'area')
df = remove_outliers(df, 'price')

print("Data after removing outliers:")
print(df)
```

```python
# Mapping categorical columns to 1/0
categorical_columns = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.map({'yes': 1, 'no': 0}))

df
```

```python
# Thanks for Ahishi
# Convert  furnishingstatus (categorical variable) into dummy/indicator variables.

# the type of furnishing can be identified with
# - status_semi-furnished, status_unfurnished
# - 0, 0  correspond to furnished
# - 0, 1  correspond to unfurnished
# - 1, 0  correspond to semi-furnished
df_dummies = pd.get_dummies(df['furnishingstatus'], prefix='status', drop_first = True)
df_dummies = df_dummies.astype(int)

df_dummies
```

```python
# Concat exisiting df with dummies
df_combined = pd.concat([df, df_dummies], axis=1)
df_combined
```

```python
# Removingfurnishingstatus
df_combined = df_combined.drop('furnishingstatus', axis=1)
df_combined
```

```python
# Shuffle the DataFrame
shuffled_df = df_combined.sample(frac=1, random_state=42).reset_index(drop=True)

# Calculate split index for 80% training and 20% testing
split_index = int(len(shuffled_df) * 0.8)

# Split the data
train_df = shuffled_df[:split_index]
test_df = shuffled_df[split_index:]
test_df
```

```python
# Min-Max Scaling function
def min_max_scale(column):
    return (column - column.min()) / (column.max() - column.min())

# Columns to scale
scale_columns = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

# Apply Min-Max scaling
train_df = train_df.copy()
train_df[scale_columns] = train_df[scale_columns].apply(min_max_scale)

test_df = test_df.copy()
test_df[scale_columns] = test_df[scale_columns].apply(min_max_scale)


print("\nTraining Data after Min-Max Scaling:")
train_df
```

```python
# Prepare the design matrix X (including bias term)
X_train = np.hstack((np.ones((train_df.shape[0], 1)), train_df.drop(columns=['price']).values))
y_train = train_df['price'].values.reshape(-1, 1)

# Calculate theta using the normal equation
theta = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train

print("\nTheta (Linear Regression Coefficients):")
print(theta)
```

```python
# Prepare the design matrix X for the test data
X_test = np.hstack((np.ones((test_df.shape[0], 1)), test_df.drop(columns=['price']).values))

# Predict the prices using the trained model
y_pred = X_test @ theta

print("\nPredicted Prices:")
print(y_pred)

# Visualizing the actual vs predicted prices
import matplotlib.pyplot as plt

plt.scatter(range(len(y_pred)), test_df['price'], color='blue', label='Actual Price')
plt.scatter(range(len(y_pred)), y_pred, color='red', label='Predicted Price')
plt.xlabel('Test Data Index')
plt.ylabel('Price (scaled)')
plt.title('Actual vs Predicted Prices')
plt.legend()
plt.show()
```

```python
# Calculate residuals
residuals = test_df['price'].values.reshape(-1, 1) - y_pred

# Residual analysis
print("\nResiduals (Actual - Predicted):")
print(residuals)

# Visualize residuals
# plt.scatter(range(len(residuals)), residuals, color='purple')
# plt.axhline(y=0, color='red', linestyle='--')
# plt.xlabel('Test Data Index')
# plt.ylabel('Residuals')
# plt.title('Residual Analysis')
# plt.show()
plt.figure(figsize=(10, 6))
plt.hist(residuals)
```

### Classification

Classification is a fundamental task in data science where the goal is to predict the categorical label of a given input based on training data. Here are some popular classification algorithms:

---
#### Logistic Regression

Logistic Regression is a linear model for binary classification that estimates the probability that a given input belongs to a certain class.

##### Logistic Function (Sigmoid Function)

The logistic function, or sigmoid function, is defined as:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

where  z  is the linear combination of the input features:

$$
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + ... + \beta_n x_n = \mathcal{w}^T \mathbf{x}
$$

##### Logistic Regression Model

The probability that an instance belongs to class 1 is given by the logistic function:

$$
P(y=1|\mathbf{x}) = \sigma(\mathcal{w}^T \mathbf{x})
$$

The predicted class label is determined by a threshold (commonly 0.5):

$$
\hat{y} = \begin{cases} 
1 & \text{if } P(y=1|\mathbf{x}) \geq 0.5 \\
0 & \text{otherwise}
\end{cases}
$$

The parameters $\mathcal{w}$ are typically estimated using Maximum Likelihood Estimation (MLE).

---

```python
import numpy as np
import matplotlib.pyplot as plt

class LogisticRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)

            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return np.array(y_predicted_cls)

# Example usage:
if __name__ == "__main__":
    # Generate synthetic data
    np.random.seed(42)
    X = np.random.randn(100, 2)
    y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

    # Split the data into training and test sets
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Initialize and train the Logistic Regression classifier
    lr = LogisticRegression(learning_rate=0.2, iterations=1000)
    lr.fit(X_train, y_train)

    # Predict on the test set
    predictions = lr.predict(X_test)

    # Calculate accuracy
    accuracy = np.sum(predictions == y_test) / len(y_test)
    print(f"Accuracy: {accuracy * 100:.2f}%")

    # Plotting the results
    plt.figure(figsize=(10, 6))

    # Plot the test data points
    plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap='viridis', marker='o', label='Test Data')

    # Plot the predictions
    plt.scatter(X_test[:, 0], X_test[:, 1], c=predictions, cmap='coolwarm', marker='x', label='Predictions')

    # Plot decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))
    Z = lr.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.2, cmap='coolwarm')

    plt.title('Logistic Regression Classifier')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.show()
```

#### Support Vector Machines (SVM)

Support Vector Machines (SVM) is a powerful classification algorithm that works by finding the hyperplane that best separates the classes in the feature space.

##### SVM Objective

The objective of SVM is to maximize the margin, which is the distance between the hyperplane and the nearest data points from each class (support vectors).

##### SVM Decision Function

The decision function for SVM is:

$$
f(\mathbf{x}) = \mathbf{w}^T \mathbf{x} + b
$$

where:
- $\mathbf{w}$ is the weight vector.
- b is the bias term.

##### Optimization Problem

The optimization problem for a linear SVM can be formulated as:

$$
\min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
$$

subject to the constraints:

$$
y_i (\mathbf{w}^T \mathbf{x}_i + b) \geq 1 \quad \forall i
$$

For non-linearly separable data, SVM can be extended using kernel functions to map the data into a higher-dimensional space where a linear separation is possible.

### Clustering

Clustering is a technique used to group similar data points into clusters. Here are some popular clustering algorithms and their applications:


#### K-means Clustering

K-means Clustering is a popular and simple iterative algorithm that partitions a dataset into K clusters, where each data point belongs to the cluster with the nearest mean.

##### K-means Algorithm

1. **Initialization**: Choose K initial centroids randomly from the dataset.
2. **Assignment**: Assign each data point to the nearest centroid, forming K clusters.
3. **Update**: Calculate the new centroids as the mean of the data points in each cluster.
4. **Repeat**: Repeat steps 2 and 3 until the centroids do not change significantly.

##### Objective Function

The objective of K-means is to minimize the sum of squared distances between data points and their corresponding cluster centroids:

$$
J = \sum_{i=1}^{K} \sum_{j=1}^{n} \| x_j^{(i)} - \mu_i \|^2
$$

where:
- K is the number of clusters.
- $ x_j^{(i)} $ is the j-th data point in the i-th cluster.
- $ \mu_i $ is the centroid of the i-th cluster.

##### Application: Optimal Distribution Center Location

K-means Clustering can be used to determine optimal locations for distribution centers. By clustering customer locations, distribution centers can be placed at the centroids of these clusters to minimize the distance to customers, reducing transportation costs and delivery times.




```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the iris dataset
iris = load_iris()
X = iris.data[:, [0, 2]]  # Selecting two dimensions: sepal length and petal length
X = StandardScaler().fit_transform(X)  # Standardize features for better convergence

# Randomly initialize centroids
k = 3
rng = np.random.default_rng(42)  # Fixed seed for reproducibility
centroids = rng.uniform(np.min(X, axis=0), np.max(X, axis=0), (k, 2))

# Function to assign clusters
def assign_clusters(X, centroids):
    distances = np.sqrt(((X[:, np.newaxis, :] - centroids) ** 2).sum(axis=2))
    return np.argmin(distances, axis=1)

# Function to update centroids
def update_centroids(X, labels, k):
    return np.array([X[labels == i].mean(axis=0) for i in range(k)])

# Set up figure
fig, ax = plt.subplots(figsize=(8, 6))
scat = ax.scatter(X[:, 0], X[:, 1], c='gray', s=50, alpha=0.7)
centroids_plot = ax.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='x')

ax.set_title('K-Means Clustering on Iris Dataset')
ax.set_xlabel('Sepal Length (standardized)')
ax.set_ylabel('Petal Length (standardized)')

# Prepare steps for animation
steps = []
current_centroids = centroids.copy()
max_iters = 10

for _ in range(max_iters):
    # Assign clusters and append the current state
    labels = assign_clusters(X, current_centroids)
    steps.append((current_centroids.copy(), labels))
    
    # Update centroids and append the updated state
    current_centroids = update_centroids(X, labels, k)
    steps.append((current_centroids.copy(), labels))

# Define cluster colors
cluster_colors = ['blue', 'green', 'orange']

def update(frame):
    centroids_frame, labels_frame = steps[frame]
    # Set data points colors
    scat.set_color([cluster_colors[i] for i in labels_frame])
    # Set centroid colors to match clusters
    centroids_plot.set_color(cluster_colors)
    centroids_plot.set_offsets(centroids_frame)
    return scat, centroids_plot

# Create animation
ani = FuncAnimation(fig, update, frames=len(steps), interval=500, blit=True)

# Save the animation to a video file
ani.save('kmeans_iris1.mp4', writer='ffmpeg', fps=2)

plt.show()
```
